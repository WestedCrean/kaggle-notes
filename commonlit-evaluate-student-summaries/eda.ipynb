```python
import pandas as pd
import datetime

# generate a list of start dates between 2020 and 2023
start_dates_range = pd.date_range(start='2020-01-01', end='2023-01-01', freq='D')
end_dates_range = pd.date_range(start='2020-01-01', end='2023-01-01', freq='D')

names = ['Adam', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'George', 'Helen', 'Iris', 'John', 'Kevin', 'Linda', 'Mary', 'Nancy', 'Olivia', 'Peter', 'Quinn', 'Robert', 'Sarah', 'Tom', 'Ursula', 'Victor', 'Wendy', 'Xavier', 'Yvonne', 'Zachary']
surnames = ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller', 'Wilson', 'Moore', 'Taylor', 'Anderson', 'Thomas', 'Jackson', 'White', 'Harris', 'Martin', 'Thompson', 'Garcia', 'Martinez', 'Robinson', 'Clark', 'Rodriguez', 'Lewis', 'Lee', 'Walker']

start_dates_range = pd.Series(start_dates_range)
end_dates_range = pd.Series(end_dates_range)
names = pd.Series(names)
surnames = pd.Series(surnames)

df = pd.DataFrame(columns=["name", 'start_date', 'end_date', 'duration'])
# pick a random start date and end date from these lists, so that the end date is always after the start date
for i in range(100):
    start_date = start_dates_range.sample().values[0]
    # pick a end_date after the start_date
    end_date = end_dates_range[end_dates_range > start_date].sample().values[0]
    # pick a random name and surname
    name = names.sample().values[0]
    surname = surnames.sample().values[0]
    # calculate the duration
    duration = (end_date - start_date)
    
    new_row = pd.DataFrame({'name': [name + ' ' + surname], 'start_date': [start_date], 'end_date': [end_date], 'duration': [duration]})
    df = pd.concat([df, new_row], ignore_index=True)
df
```
